Se aplicará el método de la transformada inversa a la función:
$$   f(x) =  \left\{ \begin{array}{lcc}  |x| & si & -1 \leq x \leq 1 \\ \\ 0 && \text{otro caso} \end{array} \right. $$

Primero, es necesario porbar que es una fdp:

*   Por se valor absoluto, se cumple que $f(x)\geq0$
*   $\int_{-1}^1|x|dx=-\int_{-1}^0 xdx+\int_0^1xdx=\frac{1}{2}+\frac{1}{2}=1$



Luego, es necesario calcular $F(x)$:
$$   F(x) =  \left\{ \begin{array}{lcc}  -\frac{x^2}{2}+\frac{1}{2} & si & -1 \leq x \leq 0 \\ \\ \frac{x^2}{2}+\frac{1}{2} & si & 0 \leq x \leq 1 \end{array} \right. $$

Este método requiere utilizar la función acumulada inversa:

$$   F^{-1}(U) =  \left\{ \begin{array}{lcc}  \sqrt{1-2u}  & si & 0 \leq U \leq 0.5 \\ \\ \sqrt{2u-1} & si & 0.5 \leq x \leq 1 \end{array} \right. $$


```python
#Se cargan las librerías
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def transf_inv (n):           #Esta función devuelve una lista con las variables aleatorias
  lista=[]                    #   que siguen la distribución f(x)
  for i in range(n):
    u=rnd.random()            # Variable aleatoria u ~ U(0,1)
    if u<=0.5:                # El condicional determina que F^-1(U) se calcula
      x=(1-2*u)**0.5
    else:
      x=-(2*u-1)**0.5
    lista.append(x)           #Se guardan los valores de X ~ f(x)
  return lista
```

```python
C=transf_inv(1000000)           #Se calculan n varibles con distribución f(x)
plt.hist(C, bins=100)          #Imprime histograma
```

```python
def acumulativa (muestra):        #Esta función devuelve las coordenadas para graficar F^-1 (x)
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=acumulativa(C)                #Se imprime F^-1{x}
plt.step(X,Y)
plt.title('F^-1 (x)')
plt.xlabel("x")
plt.ylabel("F(x)")
```
