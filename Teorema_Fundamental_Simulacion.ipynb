```python
import random as rnd
import numpy as np

def uniformes(n):
  lista=[]
  for i in range (n):
   j=1
   x=0
   for j in range (12):
    u=rnd.random()
    x+=u
   x=x-6
   lista.append(x)
  #print(lista)
  plt.hist(lista,bins=100)
  return lista

C=uniformes(100000)
```

```python
def teorema(n):
  lista=[]
  c=135/64
  for i in range (n):
    Y=rnd.random()
    u=rnd.random()
    if u <= f(Y)/c:
      lista.append(Y)
  plt.hist(lista,bins=150)
  #lista
  return lista
```

```python
C=teorema(1000000)
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(C)
plt.step(X,Y)
```
