```python
import pandas as pd  #Leer datos Excel
import numpy as np
from tabulate import tabulate  #Formato de tabla

datos=df.iloc[0:10, 0:10]  #Lee valores

# Rangos
intervalos = [0,0.2, 0.4, 0.6, 0.8, 1]
rangos = pd.cut(datos.values.flatten(), bins=intervalos)

total=rangos.size #Total de valores
fe=total/(len(intervalos)-1) #Frecuencia esperada

#Frecuencia real
conteo_rangos = rangos.value_counts()

# Crear un DataFrame a partir de los conteos
df_conteo = pd.DataFrame(conteo_rangos, columns=["f0"])
df_conteo.index.name = 'Intervalos'
df_conteo['f0-fe'] = df_conteo['f0'] - fe
df_conteo['(f0-fe)^2'] = (df_conteo['f0'] - fe)**2
df_conteo['(f0-fe)^2 / fe'] = (df_conteo['f0'] - fe)**2/fe

df_conteo = df_conteo[['f0', 'f0-fe', '(f0-fe)^2', '(f0-fe)^2 / fe']]

suma_chi=df_conteo['(f0-fe)^2 / fe'].sum()
suma_total=df_conteo['f0'].sum()

df_conteo.loc['Suma total'] = [suma_total, '', '', suma_chi]
# Imprimir el DataFrame
print(tabulate(df_conteo, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))


```
