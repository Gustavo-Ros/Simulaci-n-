# PRUEBA DE CORRIDAS

**Ejercicio 1**:
Realice una prueba de corridas con un nivel de aceptaci√≥n del 95 %
c1.PNG

```python
from scipy.stats import norm
import pandas as pd
from tabulate import tabulate

lista= [0.048, 0.591, 0.478, 0.166, 0.932, 0.926, 0.007, 0.551, 0.362, 0.533, 0.368, 0.437, 0.178, 0.411, 0.545, 0.787, 0.594, 0.086, 0.951, 0.772, 0.298, 0.199, 0.157, 0.401, 0.631]

corrida=[]
#SECUENCIA DE 1 Y 0
for i in range (1,25):
  if lista[i]>lista[i-1]:
    corrida.append(1)
  else:
    corrida.append(0)

#NUMERO DE RACHAS
rachas=1
for i in range(len(corrida)-1):
    if corrida[i] != corrida[i+1]:
        rachas+=1

#VALOR ESPERADO ,VARIANZA Y ESTADISTICO
Val_esp=(2*len(lista)-1)/3
Var=(16*len(lista)-29)/90
Estadistico=(rachas-Val_esp)/Var**0.5

print(f'Secuencia de corridas:  {corrida}')
print(f'\nN√∫mero de rachas: {rachas}\n')

#TABLA CON RESULTADOS
alfa=0.05
P_critico=norm.ppf(1 - alfa/2)
data=pd.DataFrame({'Valor esperado':[Val_esp], 'Varianza':[Var], 'Estadistico de prueba':[Estadistico],'Intervalo de confianza':[f"[{-P_critico:.4f}, {P_critico:.4f}]"]})
print(tabulate(data, headers='keys', tablefmt='fancy_grid', floatfmt=".2f", showindex=False)) #IMPRIME TABLA


if Estadistico>=-P_critico and Estadistico<=P_critico:
  print('\nSe ACEPTA la hipotesis nula: Los n√∫mero de la muestra Xi son independientes')
else:
  print('Se RECHAZA la hipotesis nula:  Los n√∫mero de la muestra Xi NO son independientes')
```
