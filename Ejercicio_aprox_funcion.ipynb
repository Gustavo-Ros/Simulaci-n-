```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(n):
  lista=[]
  for i in range(n):
    u=rnd.random()
    x=(u**0.135-(1-u)**0.135)/0.1975
    lista.append(x)
  plt.hist(lista, bins = 200)
  return lista
```

```python
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
muestra=fun(10000)
```

```python
X,Y=cdf(muestra)
plt.step(X,Y)
```
